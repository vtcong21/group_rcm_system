```python
import matplotlib.pyplot as plt

def read_metrics_from_file(file_path):
    with open(file_path, 'r') as f:
        lines = f.readlines()

    precision = float(lines[0].split(":")[1].strip())
    recall = float(lines[1].split(":")[1].strip())
    f1_score = float(lines[2].split(":")[1].strip())

    return precision, recall, f1_score

file_path_1 = 'aofram&w//data/output/online/group_2/evaluation.txt'
file_path_2 = 'my_aofram&w//data/output/online/group_2/evaluation.txt'

precision_1, recall_1, f1_score_1 = read_metrics_from_file(file_path_1)
precision_2, recall_2, f1_score_2 = read_metrics_from_file(file_path_2)

metrics = ['Precision', 'Recall', 'F1 Score']
values_1 = [precision_1, recall_1, f1_score_1]
values_2 = [precision_2, recall_2, f1_score_2]

fig, ax = plt.subplots(figsize=(10, 6))

ax.barh([metric + ' Using AOFRAM&W' for metric in metrics], values_1, color='#4CAF50', label="AOFRAM&w")

ax.barh([metric + '  Using propose AOFRAM&W' for metric in metrics], values_2, color='#FF9800', label="Propose AOFRAM&W")

ax.set_xlabel('Score')
ax.set_title('Comparison of Evaluation Metrics for Two Groups')

for index, value in enumerate(values_1):
    ax.text(value, index, f'{value:.4f}', va='center', ha='left', color='black')

for index, value in enumerate(values_2):
    ax.text(value, index + 3, f'{value:.4f}', va='center', ha='left', color='black')

ax.legend()

plt.show()
```
