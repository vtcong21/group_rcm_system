Xử lý từ file dataset => dataframe

```python
import pandas as pd
import numpy as np
import gzip
import json
```

```python
def parse(path):
    with gzip.open(path, 'rb') as g:
        for l in g:
            yield json.loads(l)
```

```python
def getDF(path):
    i = 0
    df = {}
    for d in parse(path):
        df[i] = d
        i += 1
    return pd.DataFrame.from_dict(df, orient='index')
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
```

```python
tokenizer = AutoTokenizer.from_pretrained('TungLe7661/BERT650')
model = AutoModelForSequenceClassification.from_pretrained('TungLe7661/BERT650').to(device)
```

```python
# Kiểm tra xem GPU có sẵn không
if torch.cuda.is_available():
    device = torch.device("cuda")
    print("GPU is available")
else:
    device = torch.device("cpu")
    print("GPU is not available, using CPU")

id2label={0: '1', 1: '2', 2: '3', 3: '4', 4: '5'}

trans_rating_count = 0

def sentiment_score(reviews):
    global trans_rating_count
    trans_rating_count += len(reviews)  # Tăng biến đếm số lần dự đoán theo số lượng đánh giá trong batch

    # Xử lý các đánh giá trong batch
    encoded_reviews = tokenizer(reviews, truncation=True, padding=True, return_tensors="pt")
    inputs = {key: value.to(device) for key, value in encoded_reviews.items()}  # Di chuyển dữ liệu lên GPU

    with torch.no_grad():
        output = model(**inputs)

    logits = output.logits  # Lấy logits từ output
    probabilities = torch.nn.functional.softmax(logits, dim=-1)  # Sử dụng softmax để chuẩn hóa xác suất

    # Lấy nhãn có xác suất cao nhất cho mỗi đánh giá trong batch
    predicted_labels = torch.argmax(probabilities, dim=-1)
    for i, predicted_label in enumerate(predicted_labels):

      if predicted_label.item() in id2label:
          print(f"\nPredicted Label for review {trans_rating_count - len(reviews) + i + 1}: {id2label[predicted_label.item()]}")
      else:
          print(f"\nPredicted Label for review {trans_rating_count - len(reviews) + i + 1}: Unknown label ({predicted_label.item()})")

    return [int(label.item())+1 for label in predicted_labels]



```

```python
sentiment_score(df_review['reviewText'].iloc[1])
```

```python
merged_df = pd.merge(df_meta_data[['asin']], df_review[['reviewerID', 'reviewText', 'overall', 'asin']], on='asin', how='left')
merged_df = merged_df[['overall', 'reviewerID', 'asin', 'reviewText']]
```

```python
merged_df
```

```python
# Thêm cột 'modelRating' vào DataFrame
merged_df['modelRating'] = merged_df['reviewText'].apply(lambda x: sentiment_score([x]) if pd.notna(x) and x != '' else 0)
```

```python
# Lọc các ký tự không hợp lệ khỏi cột 'reviewText'
merged_df.loc[:, 'reviewText'] = merged_df['reviewText'].apply(lambda x: ''.join([c for c in str(x) if c.isprintable()]))
# Loại bỏ dấu ngoặc vuông trong cột 'modelRating'
merged_df.loc[:, 'modelRating'] = merged_df['modelRating'].apply(lambda x: x[1:-1] if isinstance(x, str) and x.startswith('[') and x.endswith(']') else x)
# Lưu DataFrame vào file Excel

merged_df.to_excel('/content/drive/MyDrive/Colab Notebooks/data/final/All_Beauty.xlsx', index=False)
```
